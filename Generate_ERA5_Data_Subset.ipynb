# Generate an ERA5 sample from GCP's GS storage

```python
# Generate an ERA5 sample from GCP's GS storage
import xarray as xr
import gcsfs

## Load Zarr via Xarray
# Note: Does not extract attibutes UNLESS you declare the variable and then, test[VARIBALE_NAME].attrs
gs_url = "gs://gcp-public-data-arco-era5/ar/1959-2022-1h-360x181_equiangular_with_poles_conservative.zarr"
chunk_sz = 48
gcp_ar_era5_subset = xr.open_zarr(gs_url, 
                    chunks={'time': chunk_sz},
                    consolidated=True)

# Select data for a specific date range
start_date = '2020-12-31'
end_date = '2021-02-01'
gcp_ar_era5_subset = gcp_ar_era5_subset.sel(time=slice(start_date, end_date))
gcp_ar_era5_subset 
```

```python
# Save ERA5 data subset to local 
gcp_ar_era5_subset.to_zarr('./sample_data/gcp_era5_subset.zarr')
```
